# Pattern classification with missing data: a review
*Garca-Laencina, P.J., Sancho-Gmez, JL. & Figueiras-Vidal, A.R. Pattern classification with missing data: a review. Neural Comput & Applic 19, 263282 (2010). https://doi.org/10.1007/s00521-009-0295-6*

- Is about pattern classification (supervised)
- Problems: handling missing values and pattern classification
- Missing values can be handled by:
    - Removing the samples with missing values
    - Imputing or estimating the missing values
    - Model-based procedure (model data distribution, e.g. EM)
    - Using a classifier that can handle missing values
    - [image](https://link.springer.com/article/10.1007/s00521-009-0295-6/figures/1)
- Way of handling often depends on why data is missing (p.4)
- Missing values can be (p(M|X0, Xm, Xi):
    - Missing completely at random (MCAR, p(M|Xi))
    - Missing at random (MAR, p(M|X0, Xi)): pattern of data missingness is traceabl or predictable from other variablesin the database
    - Not missing at random (NMAR): Depends on the value of the missing data -> no general method to handle
- Ignorable (MCAR and MAR): ignore reasons for missingness  in analysis
- Relation to prediction target
    - Informative: learn from missing values
    - Non-informative: imputation
- Impution methods (statistical):
    - Mean imputation: mean of observed values (similar: class-conditional: mean of observed values in same class); ignores correlation, under represents data variety
    - Regression imputation: train regression model to predict missing values on Xo (multivariate regression model)
    - Hot deck imputation: use values from similar samples (e.g. k=1-nearest neighbors)(similar: cold deck imputation: use values from other data sources)
    - Multiple imputation: replace with set of multiple values, leading to muliple complete datasets, average parameter estimats
- Imputation methods (machine learning):
    - K-nn -> Mean/mode, distane-weighted also possible
        - Distnce function important: Euclidean (HEOM), range normalized? ...
    - Self-organizing maps (SOM): map to lower dimension, not entirely clear how it works
    - MLP: train on only complete data, one MLP per possible missing data combination, predict missing values
    - Recurrent neural network (RNN): feedback into input to estimate missing data
    - Auto-assocative neural network (AANN): completely connected neurons, train on complete data
    - Multi task learning (MTL): solve main task and extra tasks together
